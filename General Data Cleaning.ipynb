# General Data Cleaning


## Data import

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("melb_data.csv")
data
```

```python
data.YearBuilt
```

# Address missing values

```python

missing_value_num = data.YearBuilt.isnull().sum()
missing_value_num
```

```python
total_value_num = np.product(data.YearBuilt.shape)
missing_value_num/total_value_num
```

```python
# the easiest way

cleaned_data = data.dropna(subset = ['YearBuilt'])
cleaned_data
```

```python
from sklearn.impute import SimpleImputer
```

```python
# imputation

imputer = SimpleImputer(strategy = 'mean')
year_built = imputer.fit_transform(pd.DataFrame(data.YearBuilt))
print(year_built)
```

```python
# expanded imputation

data.YearBuilt_Missing = data.YearBuilt.isnull()
print(data.YearBuilt_Missing)
```

# Scaling and Normalization

```python
from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(data.Price)
scaled_price = minmax_scaling(data.Price, columns = [0])
scaled_price
```

```python
normalized_price = stats.boxcox(data.Price)[0]
print(normalized_price)
```

# Character Encoding

```python
string = "This is the euro symbol: €"
string = string.encode("utf-8")
print(string)
```

```python
string = string.decode("utf-8")
print(string)
```

```python
string = string.decode("ascii")
print(string)
```

```python
string = string.decode("ascii", errors = 'replace')
print(string)
```
